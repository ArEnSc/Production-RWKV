```python
# reloads notebook so you don't have to reset it when the file changes
%load_ext autoreload
%autoreload 2
# only run once to interop with the files
import sys
class NotebookSourceInterop:
    def __init__(self) -> None:
        self.file_pathes = set()
    def add_pathes(self,path:str):
        self.file_pathes.add(path)
    def display_pathes(self):
        print(sys.path)
    def _remove_pathes(self):
        for i in self.file_pathes:
            try:
                sys.path.remove(i) 
            except Exception as e:
                print(e)
    def _restore_pathes(self):
        for i in self.file_pathes:
            sys.path.append(i)
    def refresh(self):
        self._remove_pathes()
        print(f"Restoring: {sys.path}")
        self._restore_pathes()
        print(f"Restored: {sys.path}")
nbi = NotebookSourceInterop()
nbi._remove_pathes()
nbi.add_pathes(path="./RWKV-LM/RWKV-v4neo/")
nbi.add_pathes(path="../src/")
nbi.refresh()
```

```python
from rwkvtokenizer import RWKVTokenizer
from rwkvrnnmodel import RWKVRNN4NeoForCausalLM
```

```python
tokenizer = RWKVTokenizer.from_file()
model = RWKVRNN4NeoForCausalLM.from_pretrained()
```

```python
context = "\nIn a shocking finding, scientist discovered a herd of dragons living in a remote, previously unexplored valley, in Tibet. Even more surprising to the researchers was the fact that the dragons spoke perfect Chinese."
context_input_ids = tokenizer.encode(context).ids
```

```python
model._warmup_with_context(context=context_input_ids,save_path=None)
```

```python
def callback(ind):
    index = tokenizer.decode([ind],skip_special_tokens=False)
    print(ind)
    # print(repr(index),end="")

ctx = model.generate(inputs=context_input_ids,streaming_callback=callback,max_length=512)

print(repr(tokenizer.decode(ctx,skip_special_tokens=False))) # cpu 3 tokens a second
```
